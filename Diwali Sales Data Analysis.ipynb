```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importdata
df = pd.read_csv(r'/Users/poonampetal/Downloads/Python_Diwali_Sales_Analysis/Diwali Sales Data.csv', encoding = "unicode_escape")
```

```python
#check data
df.shape
```

```python
#check data details
df.head()
```

```python
#delete unwanted columns
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True )
```

```python
df.info()
```

```python
#check null values
df.isnull().sum()
```

```python
#delete null values
df.dropna(inplace = True )
```

```python
#change data type 
df['Amount'] = df['Amount'].astype('int')
```

```python
#check data type 
df['Amount'].dtype
```

```python
#rename column
df.rename(columns = {'Marital_Status' : 'Shaadi'})
```

```python
df['Marital_Status'] = df['Marital_Status'].astype('boolean')
```

```python
df.info
```

```python
Exploratory Data Analysis
```

```python
Gender
```

```python
ax = sns.countplot( x = 'Gender',data = df )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender = df.groupby(['Gender'],as_index=False)['Amount'].sum()
print(sales_gender)
```

```python
ax = sns.barplot(x='Gender', y= 'Amount', data = sales_gender)
```

```python
#from the above grabph we can see that most of the buyers are female even the purchasing power of females is greater than male.
```

```python
#Age
```

```python
ax = sns.countplot( x = 'Age Group',data = df, hue = 'Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum()

ax = sns.barplot(x='Age Group', y= 'Amount', data = sales_age)
```

```python
#From above graph we can see that most of the buyers are from age group between 26 - 35 and they are mostly single.
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().head(8).sort_values(by='Orders',ascending = False)
sns.set(rc={'figure.figsize':(13,5)})
ax = sns.barplot(data = sales_state,x='State', y= 'Orders')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().head(8).sort_values(by='Amount',ascending = False)
sns.set(rc={'figure.figsize':(13,5)})
sns.set(rc={'figure.figsize':(13,5)})
ax = sns.barplot(data = sales_state, x='State', y='Amount')

```

```python
# From the above graph we can see that most of ther orders and total sales/amount are from Karnataka, Delhi and Andhra Pradesh.
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(10,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
*From above graphs we can see that most of the buyers are single (women) and they have high purchasing power*
```
